```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.impute import SimpleImputer
from sklearn.impute import KNNImputer
from sklearn.impute import SimpleImputer
```

```python
#load dataset
data = pd.read_csv('2019_SANJUAN.csv')
```

```python


# Load your dataset
file_path = '2019_SANJUAN.csv'  # Replace with your file path
df = pd.read_csv(file_path)

# Separate numerical and categorical columns
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

print("Missing values before imputation:")
print(df.isnull().sum())

# 1. KNN Imputation for numerical columns
knn_imputer = KNNImputer(n_neighbors=5)  # You can change the number of neighbors
df[numerical_cols] = knn_imputer.fit_transform(df[numerical_cols])

# 2. Simple Imputer for categorical columns (using 'most_frequent')
simple_imputer = SimpleImputer(strategy='most_frequent')
df[categorical_cols] = simple_imputer.fit_transform(df[categorical_cols])

# Save the updated DataFrame back to a CSV file
output_path = 'updated_file.csv'
df.to_csv(output_path, index=False)

print("Missing values after imputation:")
print(df.isnull().sum())
print(f"Updated file saved to {output_path}")
```

```python
#Preprocess dataset
label_encoders = {}
for col in ['barangay', 'typeofPlace', 'modus', 'casestatus', 'stageoffelony']:
    label_encoders[col] = LabelEncoder()
    data[col] = label_encoders[col].fit_transform(data[col])
```

```python
# Extract date/time features
data['hourReported'] = pd.to_datetime(data['timeReported'], format='mixed').dt.hour
data['hourCommitted'] = pd.to_datetime(data['timeCommitted'], format='mixed').dt.hour
```

```python
features = ['barangay', 'typeofPlace', 'hourReported', 'hourCommitted', 'lat', 'lng', 'stageoffelony']
X = data[features]
y = data['offense']  # Target: predicting offense type
```

```python
# Encode target variable
y = LabelEncoder().fit_transform(y)
```

```python
# Step 3: Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Step 4: Train Naive Bayes
model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
# Step 5: Predictions
y_pred = model.predict(X_test)
```

```python
# Step 6: Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
```

```python
print(f"Accuracy: {accuracy}")
print(conf_matrix)
```
